### Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

- Scatter plot represents individual data points as dots on a 2D plane, useful for visualizing relationships between two variables.

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0,2,50)
y = 3 + np.random.normal(0,2, len(x))
plt.scatter(x,y)
plt.xlabel('X data')
plt.ylabel('Y data')
plt.title('Data Information')
```

### Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

- The 'subplot()' functio in matplotlib is used to create multiple plots within a single figure, allowing to arrange plots in a grid like structure

```python
import numpy as np
import matplotlib.pyplot as plt

# data for line 1
x1 = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([0, 100, 200, 300, 400, 500])

# data for line 2
x2: x = np.array([0, 1, 2, 3, 4, 5]) 
y2 = np.array([50, 20, 40, 20, 60, 70])

# data for line 3
x3 = np.array([0, 1, 2, 3, 4, 5]) 
y3 = np.array([10, 20, 30, 40, 50, 60])

# data for line 4
x4 = np.array([0, 1, 2, 3, 4, 5]) 
y4 = np.array([200, 350, 250, 550, 450, 150])

# creating 2x2 grid of subplots
plt.subplot(221)
plt.plot(x1,y1, label = 'Line 1')
plt.title('Line 1')

plt.subplot(222)
plt.plot(x2,y2, label = 'Line 2')
plt.title('Line 2')

plt.subplot(223)
plt.plot(x3,y3, label = 'Line 3')
plt.title('Line 3')

plt.subplot(224)
plt.plot(x4,y4, label = 'Line 4')
plt.title('Line 4')

plt.tight_layout()
plt.show()
```

### Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

- A bar plot is a type of data visualization that represents categorical data with rectangular bars. Each bar typically corresponds to a category, and the length or height of the bar represents the value or frequency of that category. Bar plots are commonly used to compare different categories or show the distribution of a dataset across discrete categories.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company, profit)
plt.xlabel("Company")
plt.ylabel("Profit")
plt.show()
```

```python
plt.barh(company, profit)
plt.xlabel("Company")
plt.ylabel("Profit")
plt.show()
```

### Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

- A box plot is a graphical representation of the distribution of a dataset, providing a summary of its central tendency, dispersion, and potential outliers. It displays the data's median, quartiles, and extreme values, making it easy to compare the distribution of multiple datasets.

Box plots are used for various purposes, such as:

- Visualizing Data Distribution: Box plots provide a quick visual summary of how data is spread out, allowing you to observe symmetry, skewness, and identify potential outliers.

- Comparing Datasets: Box plots are helpful when comparing multiple datasets side by side, enabling you to understand differences in their central tendency and spread.

- Identifying Outliers: Box plots can help detect potential outliers in a dataset, which are data points that fall far away from the typical range.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data = [box1, box2]
plt.boxplot(data)
plt.show()
```
